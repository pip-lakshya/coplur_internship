```python
#Question 1: 1) Write a python program that takes in a student name, class.
#  It should also take in five subject marks of the students and find the total mark and percentage. Display a result in such a way that their name, class,  and percentage are printed.
students=[]
stud={"name":None,"class":None,"marks":[],"percentage":None,"total":0}
f=0
while True:
    f=int(input("press 1 if you want to enter a student"))
    if f!=1: break
    stud["name"]=input("Enter name of student")
    stud["class"]=input("Enter class")
    stud["marks"], stud["total"]=[],0
    for i in range(5):
        mark=int(input(f"Enter marks in subject {i+1}:"))
        stud["marks"].append(mark)
        stud["total"]+=stud["marks"][i]
    stud["percentage"]=(stud["total"]/500)*100
    #4) In your last program where you find the total and percentage of a student's marks of 5 subject, find the grade of the student using conditional statement. Eg. grade 'A' if percentage is greator than or equals to 60, 'B' for  percentage is greator than or equals to 50 and less than 60,  'C' for  percentage is greator than or equals to 40 and less than 50,  'D' for  percentage is greator than or equals to 33 and less than 40, otherwise 'Fail'
    if(stud["percentage"]>=60):
        stud["grade"]="A"
    elif(stud["percentage"]>=50 and stud["percentage"]<60 ):
        stud["grade"]="B"
    elif(stud["percentage"]>=40 and stud["percentage"]<50):
        stud["grade"]="C"
    elif(stud["percentage"]>=33 and stud["percentage"]<40):
        stud["grade"]="D"
    else:
        stud["grade"]="Fail"
    print(stud)
    students.append(stud)
print(students)
```
