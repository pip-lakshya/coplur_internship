```python
l=[]
while(True):
    print("""1.add element \n2. remove element \n3. display first three elements \n4.display last three elements \n5. print \n6. exit\n\n""")
    ch= int(input("enter your choice: "))
    if ch==1:
        e=input("enter an element to be added: ")
        l.append(e)
    elif ch==2:
        if l.__len__()==0:
            print("list is empty")
        else:
            l.pop(0)
    elif ch==3:
        if(len(l)>=3):
            print("The top 3 elements are:")
            for i in range(3):
                print(l[i])
            print("\n")
        else:
            print("Que smaller than 3 elements")
    elif ch==4:
        if(len(l)>=3):
            print("The bottom 3 elements are:")
            for i in range(-1,-4,-1):
                print(l[i])
            print("\n")
        else:
            print("Que smaller than 3 elements")
    elif ch==5:
        if(l.__len__==0):
            print("empty queue")
        else:
            print("Elements in queue are:")
            for i in l:
                print(i)
            print("\n")
    elif ch==6:
        break
    else:
        print("invalid choice")

    
```
